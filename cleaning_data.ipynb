```python
#add the pandas dependency
import pandas as pd
```

```python
#files to load
file_to_load = "Resources/missing_grades.csv"

#read the csv into a dataframe
missing_grade_df = pd.read_csv(file_to_load)

missing_grade_df
```

```python
#Option 1 - Do nothing; the NaNs will not be considered in sums or averages and therefore the missing values have no impact. 
#If we multiply or divide with a row with NaN, the answer will be NaN. 
#This can cause problems if we need those answers for the rest of our code.

#Option 2 - Drop the row; drop the rows where there are NaNs. We will remove the rest of the data with the NaN, too.
#This can cause problems later if there is other data in the other columns that we need. 
#Dropping rows can affect the story we are trying to tell with the data. Ask yourself 2 questions before removing rows:
    #1. How much data would be removed if NaNs are dropped?
    #2. How would this impact the analysis? 

#Drop the NaNs.
missing_grade_df.dropna()
```

```python
#Option 3 - Fill in the Row; fill in all the NaNs with a value
#must be used with caution since it can impact arithmetic calculations (like with 0), if you fill in, the values must be
#carefully considered for every downstream analysis you perform.

#fill in the empty rows with "85".
missing_grade_df.fillna(85)
```
